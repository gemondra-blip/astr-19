```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_variates = np.random.gamma(shape=2.0, scale=2.0, size=1000)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(random_variates, bins=100, color='coral', edgecolor='black', alpha=0.8)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Variates (Gamma Distribution)')
```
