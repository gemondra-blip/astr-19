```python
import numpy as np
```

```python
def sin(x):
    return np.sin(x)
'''
function: sin(x)

inputs: x, data type is float or int

output: the value of sin(x)
'''
```

```python
def cos(x):
    return np.cos(x)
'''
function: cos(x)

inputs: x, data type is float or int

output: the value of cos(x)
'''
```

```python
def CosVsSin(x):
    for i in range(len(x)):
        print(f'{cos(x[i])} vs {sin(x[i])}')
'''
Function: CosVsSin()

inputs:
    x ; Data type float/int
Outputs:
    It calls a print that then calls cos(x) and sin(x) with the value CosVsSin was called with
'''
```

```python
def XvsSINvsCOS(x):
    for i in range(10):
        print(f"{x[i]} vs {cos(x[i])} vs {sin(x[i])}")
```

```python
def Xtabulator():
    
    x = np.linspace(0,2*np.pi, 1000)

    return x

    
    
    '''
Creating a range between 0 -- 2pi, then filling a array with evenly 
spaced steps summing to 1000 steps withing the range
    '''
    
    
```

```python
x = Xtabulator()
```

```python
CosVsSin(x)
```

```python
XvsSINvsCOS(x)
```
