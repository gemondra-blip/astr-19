```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)

def Exp(a = 1, b = 1, x = 1):
    a = a
    b = b
    x = x
    return a*np.exp(-b*x)
###a = float(input('Give a value for a: '))
###b = float(input('Give a value for b: '))

a = 5
b =2
```

```python
fig = plt.figure()
ax = plt.subplot()
y = Exp(a = a, b = b, x = x)
ax.plot(x, y)
ax.plot(x, x)
ax.set_xlabel("X ranging from 0 - 2Pi", rotation = 0)
ax.set_ylabel( 'Y', rotation = 0)
ax.set_title("x vs y")
ax.legend(['y','x'])
plt.savefig('prompt3_curvefit.pdf')
plt.show()
```
