```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2 * np.pi, 100)
a = 5.0
b = 0.5
y = a * np.exp(-b * x)
```

```python
m = (0.2 - 2.0) / (2 * np.pi - 0)
y_err = m * x + 2.0

y_scattered = y + np.random.normal(0, y_err)

plt.figure(figsize=(8, 6))

plt.errorbar(x, y_scattered, yerr=y_err, fmt='o', color='purple', ecolor='lightgray', 
             elinewidth=1.5, capsize=3, label='Data with varying uncertainty')

plt.xlabel('x (radians)')
plt.ylabel('y (scattered)')
plt.title('Scatter Data with Linearly Decreasing Uncertainty')
plt.legend()

plt.savefig('prompt2_errorbar.pdf')
plt.show()
```
